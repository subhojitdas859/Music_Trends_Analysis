# Analysis

### TOP 10 least popular songs


#### Find the top 10 least popular songs

```python
# Sort the dataframe by popularity in ascending order, fetch the first 10 rows
sorted_df = tracks_df.sort_values('popularity', ascending=True).head(10)
sorted_df
```

### Top 10 songs with popularity > 90

#### Find the 10 most popular songs with popularity > 90

```python
most_popular = tracks_df.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
# Here, I will change the current index to release date
tracks_df.set_index('release_date',inplace=True)
tracks_df.index = pd.to_datetime(tracks_df.index)
tracks_df.head()
```

```python
# Change the info
tracks_df.info()
```

```python
# Convert duration in ms to sec
tracks_df['duration'] = tracks_df['duration_ms'].apply(lambda x:round(x/1000))
#Drop the duration_ms column
tracks_df.drop('duration_ms', inplace=True, axis=1)
```

```python
# Display all the columns names
tracks_df.columns
```

## Drop the unnecessary column

```python
corr_df = tracks_df.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
```

#### Find the correlation map

#### Using the heatmap function from the seaborn library, we will try to plot the correlationship heatmap between varaibles.

```python
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1,vmax=1, center=0,cmap="inferno",lw=0.75, linecolor="Black")
heatmap.set_title("Correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

### From the above correlation heatmap,we can see the following major inference:
* Loudness and energy has a positive correlation of 0.8. So if the loudness of the track is high, that means the energy of the song is also high.
* Also, we see that valence and danceability has a positive correlation of 0.5.
* Popularity and acousticness has a negative correaltion of 0.4, if a track is higly popular then the acousticness of the track is low and vice versa.

#### Now, let us take a sample data set from the tracks_df dataframe and see the above inference in the visualization.

```python
sample_df = tracks_df.sample(int(0.004*len(tracks_df)))
sample_df
```

#### Since there was positive correlation between loudness and energy, plot a graph for between them.


### Regression line

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs Energy Correlation")
```

#### So, it is evident from the regression plot that is the loudness of the track is hight then, energy is also hight.

### Regression plot between popularity and acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="r").set(title="Popularity vs Acousticness Correlation")
```

### Like mentioned earlier, we can see that higher the popularity of the song, lower the acousticness and lower the popularity higher the acoustiness.

### Now, let us find the total number of songs each year since 1992.

```python
# Create a new column called 'dates' and get the year data in a variable called 'years'
tracks_df['dates'] = tracks_df.index.get_level_values('release_date')
tracks_df.head()
tracks_df.dates = pd.to_datetime(tracks_df.dates)
years = tracks_df.dates.dt.year
```

```python
# See the years values
years.head()
```

### As a last analysis step, I will now consider the total number of songs each year since 1922.

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

##### Thus, from the above plot, we can see that the total songs in from 1922 to 1960 has seen a increasing trend. However, there was a sudden drop in the number of songs during 1974 and then the number of songs increased the next few years. We can also see that from 2000 to 2020 there is a positive trend in the number of songs, but in the year 2021 there was a drop in the 2021.
